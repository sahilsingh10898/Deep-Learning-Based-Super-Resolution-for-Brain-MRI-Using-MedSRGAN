```python
def discriminator_model(hr_img_shape=(None,None,3),lr_img_shape=(None,None,3)):

    
    input_hr=layers.Input(shape=hr_img_shape)
    input_lr=layers.Input(shape=lr_img_shape)

    # for hr images
    x_hr=Conv2D(64,kernel_size=3,strides=1,padding='same')(input_hr)
    x_hr=LeakyReLU(alpha=0.2)(x_hr)

    x_hr=Conv2D(64,kernel_size=3,strides=2,padding='same',name="conv_0")(x_hr)
    x_hr=BatchNormalization()(x_hr)
    x_hr=LeakyReLU(alpha=0.2)(x_hr)

    x_hr=Conv2D(128,kernel_size=3,strides=1,padding='same')(x_hr)
    x_hr=BatchNormalization()(x_hr)
    x_hr=LeakyReLU(alpha=0.2)(x_hr)

    x_hr=Conv2D(128,kernel_size=3,strides=2,padding='same',name="conv_1")(x_hr)
    x_hr=BatchNormalization()(x_hr)
    x_hr_f=LeakyReLU(alpha=0.2)(x_hr)

    # for low res image

    x_lr=Conv2D(64,kernel_size=3,strides=1,padding='same')(input_lr)
    x_lr=LeakyReLU(alpha=0.2)(x_lr)

    x_lr=Conv2D(128,kernel_size=3,strides=1,padding='same')(x_lr)
    x_lr=BatchNormalization()(x_lr)
    x_lr_f=LeakyReLU(alpha=0.2)(x_lr)

    concat_layer=Concatenate()([x_hr_f,x_lr_f])

    #feature Extraction Layers

    x=Conv2D(256,kernel_size=3,strides=1,padding='same')(concat_layer)
    x=BatchNormalization()(x)
    x=LeakyReLU(alpha=0.2)(x)

    x=Conv2D(256,kernel_size=3,strides=2,padding='same',name="conv_2")(x)
    x=BatchNormalization()(x)
    x=LeakyReLU(alpha=0.2)(x)

    x=Conv2D(512,kernel_size=3,strides=1,padding='same')(x)
    x=BatchNormalization()(x)
    x=LeakyReLU(alpha=0.2)(x)

    x=Conv2D(512,kernel_size=3,strides=2,padding='same',name="conv_3")(x)
    x=BatchNormalization()(x)
    x=LeakyReLU(alpha=0.2)(x)

    x=Conv2D(1024,kernel_size=3,strides=1,padding='same')(x)
    x=BatchNormalization()(x)
    x=LeakyReLU(alpha=0.2)(x)

    x=Conv2D(1024,kernel_size=3,strides=2,padding='same',name="conv_4")(x)
    x=BatchNormalization()(x)
    x=LeakyReLU(alpha=0.2)(x)

    # fully connected Layers

    x=GlobalAveragePooling2D()(x)
    x = Dense(100, activation='leaky_relu')(x)
    x = Dense(1, activation='sigmoid')(x)
    
    model=Model([input_hr,input_lr],x)

    return model
    
```
